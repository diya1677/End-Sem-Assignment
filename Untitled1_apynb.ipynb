# **Movie** Data Analysis

```python
import pandas as pd
import ast
from collections import Counter
df = pd.read_excel("/content/imb.121.xlsx")

print("Dataset Loaded Successfully!\n")
```

Q1.Which movie made the highest profit?who were its producer and director?Identify that film.

```python
def ext_role(crew_str, role):
    try:
        crew_list = ast.literal_eval(crew_str)
        for member in crew_list:
            if member.get("job") == role:
                return member.get("name")
    except (ValueError, SyntaxError):
        return None
    return None

def ext_names(cast_str):
    names = []
    try:
        cast_list = ast.literal_eval(cast_str)
        for member in cast_list:
            if "name" in member:
                names.append(member["name"])
    except (ValueError, SyntaxError):
        return []
    return names

df["profit"] = df["revenue"] - df["budget"]

df["ROI"] = df.apply(lambda x: (x["profit"] / x["budget"])
    if x["budget"] > 0 else 0,
    axis=1
)
max_profit_movie = df.loc[
    df["profit"].idxmax()
]

print("\nHighest Profit Movie:")
print(max_profit_movie["title"])

print("\nProducer:")
print(ext_role(max_profit_movie["crew"],"Producer"))

print("\nDirector:")
print(ext_role(max_profit_movie["crew"],"Director"))
print("\nActors:")
print(ext_names(max_profit_movie["cast"]))
```

Q2. Which language has the highest average ROI (return on investment)?

```python
language_roi = (df[df["budget"] > 0].groupby("original_language")["ROI"].mean().sort_values(ascending=False))
print("\nLanguage with Highest ROI:")
print(language_roi.head())
```

3. Find out the unique genres of movies in this dataset.

```python
unique_genres = set()

for g in df["genres"]:
    unique_genres.update(ext_names(g))
print("\nUnique Genres:")
print(sorted(unique_genres))
```

Q4. Make a table of all the producers and directors of each movie. Find the top 3 producers who have produced movies with the highest average RoI?

```python
producer_data = []

for _, row in df.iterrows():
  producer_name = ext_role(row["crew"],"Producer")
  producers_list = [producer_name] if producer_name is not None else []
  for p in producers_list:
    producer_data.append(
      {
        "Producer": p,
        "ROI": row["ROI"]
      }
    )

producer_df = pd.DataFrame(
    producer_data
)

top_producers = (
    producer_df
    .groupby("Producer")["ROI"]
    .mean()
    .sort_values(ascending=False)
    .head(3)
)

print("\nTop 3 Producers by ROI:")
print(top_producers)
```

Q5. Which actor has acted in the most number of movies? Deep dive into the movies, genres and profits corresponding to this actor.

```python
actor_counter = Counter()

for c in df["cast"]:

    actor_counter.update(
        ext_names(c)
    )
top_actor, count = (
    actor_counter.most_common(1)[0]
)
print("\nActor with Most Movies:")
print(top_actor)
print("Number of movies:", count)
actor_movies = []

for _, row in df.iterrows():

    cast_list = ext_names(
        row["cast"]
    )

    if top_actor in cast_list:

        genres = ext_names(
            row["genres"]
        )

        actor_movies.append(
            {
                "Movie": row["title"],
                "Genres": genres,
                "Profit": row["profit"]
            }
        )

actor_movies_df = pd.DataFrame(
    actor_movies
)

print("\nMovies by Actor:")
print(actor_movies_df.head())
```
